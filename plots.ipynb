```python
import plotly as py
import plotly.figure_factory as ff
import pandas as pd
import plotly.graph_objects as go
import pyproj
from datetime import datetime, timedelta, timezone
py.offline.init_notebook_mode(connected=True)
from astropy.time import Time

pos_header= ['date','time','x','y','z','Q','ns','sdn','sde','sdu','sdne','sdeu','sdun','age','ratio']
track_header=['YY','DOY','Seconds','Latitude','Longitude','Height','SigN','SigE','SigH','RMS','#','Atm','+-','Fract_DOY','Epoch','#BF','NotF','F']
pride_header=['Mjd','Sod', 'X','Y','Z','Latitude','Longitude','Height','Nsat', 'G', 'R', 'E', 'C2', 'C3', 'J','PDOP']

xyzlla = pyproj.Transformer.from_crs(
    {"proj":'geocent', "ellps":'WGS84', "datum":'WGS84'},
    {"proj":'latlong', "ellps":'WGS84', "datum":'WGS84'},
    )
```

```python
pos = pd.read_csv(open("19177-Velsk-Arch.pos", errors='replace'),sep='\s+',comment='%', skiprows=481,names=pos_header)
pos[['lat','lon','alt']] = pos.apply(lambda row: xyzlla.transform(row.x,row.y,row.z,radians=False), axis='columns', result_type='expand')
pos['dt'] = pos.apply(lambda row: datetime.strptime(row.date+' '+row.time,'%Y/%m/%d %H:%M:%S.%f'), axis='columns')
pos = pos.set_index('dt')
pos = pos["2019-06-26 09:45:00.000":"2019-06-26 10:05:00.000"]
pos.head()
```

```python
track = pd.read_csv(open("track1-aug-POS.GEOD.trac.LC", errors='replace'),sep='\s+',comment='*', skiprows=0,index_col=False,names=track_header)
track['dt'] = track.apply(lambda row: datetime.strptime(str(row.YY),'%Y') + timedelta(days=row.DOY-1) + timedelta(seconds=row.Seconds), axis='columns')
track = track.set_index('dt')
track = track["2019-06-26 09:45:00.000":"2019-06-26 10:05:00.000"]
track.head()
```

```python
pride = pd.read_csv(open("kin_2019177_r177", errors='replace'),sep=' ', skiprows=43,index_col=False,names=pride_header,on_bad_lines='skip',skipinitialspace=True)
pride['dt'] = pride.apply(lambda row: Time(row.Mjd+2400000.5,format='jd').to_datetime()+timedelta(seconds=row.Sod), axis='columns')
pride = pride.set_index('dt')
pride = pride["2019-06-26 09:45:00.000":"2019-06-26 10:05:00.000"]
pride.head()
```

```python
trace_z_alt1= go.Scatter(x=pos.index,y=pos.alt,name="rtklib")
trace_z_alt2 = go.Scatter(x=track.index,y=track.Height,name="TRACK")
trace_z_alt3 = go.Scatter(x=pride.index,y=pride.Height,name="PRIDE")
```

```python
data = [trace_z_alt1,trace_z_alt2,trace_z_alt3]
layout = go.Layout(title=" ")#,xaxis_range=[0,11500],yaxis_range=[420,530])

fig = go.Figure(data=data, layout=layout)
py.offline.iplot(fig)
```
